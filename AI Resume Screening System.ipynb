```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Role": [
        "Data Analyst",
        "Data Scientist",
        "Machine Learning Engineer",
        "Business Analyst",
        "Software Developer"
    ],
    "Required_Skills": [
        "python, sql, excel, statistics, power bi",
        "python, machine learning, statistics, data visualization",
        "python, machine learning, deep learning, tensorflow",
        "excel, sql, communication, business analysis",
        "java, oops, dsa, sql"
    ]
}

df = pd.DataFrame(data)
df
```

```python
user_input = input("Enter your skills (comma separated): ")

user_skills = [skill.strip().lower() for skill in user_input.split(",")]

print(user_skills)
```

```python
results = []

for index, row in df.iterrows():

    role = row["Role"]
    required_skills = [s.strip().lower() for s in row["Required_Skills"].split(",")]

    matched = [s for s in user_skills if s in required_skills]
    score = len(matched)

    missing = [s for s in required_skills if s not in user_skills]

    if len(required_skills) > 0:
        percent = (score / len(required_skills)) * 100
    else:
        percent = 0

    results.append([role, score, round(percent,2), missing])

results.sort(key=lambda x: x[1], reverse=True)
```

```python
print("\n🔥 Top 3 Job Matches:\n")

for r in results[:3]:
    print("Role:", r[0])
    print("Match Score:", r[1])
    print("Match %:", r[2], "%")
    print("Missing Skills:", ", ".join(r[3]))
    print("-" * 40)

best = results[0]

print("\n🏆 Best Job Role:", best[0])
print("Match %:", best[2], "%")
```

```python
import pdfplumber

file_name = list(uploaded.keys())[0]

text = ""

with pdfplumber.open(file_name) as pdf:
    for page in pdf.pages:
        page_text = page.extract_text()
        if page_text:
            text += page_text + " "

text = text.lower()

print("\n--- EXTRACTED TEXT ---\n")
print(text)
```

```python
skills_list = [
    "python", "sql", "excel", "power bi",
    "machine learning", "deep learning",
    "statistics", "java", "dsa", "oops",
    "data visualization"
]

extracted_skills = []

for skill in skills_list:
    if skill in text:
        extracted_skills.append(skill)

print("\n🔥 Extracted Skills:", extracted_skills)
```
